## 2. Drop columns not used

```python
df.columns
```

```python
df.drop(["FIPS code[3]", "County seat[4]", "Origin[5]", "Map"], axis='columns', inplace=True)
df.head()
```

## 3. Rename columns

```python
df.columns
```

```python
cols_to_rename = {
                   'County':'County_Name_Full',
                   'Est.[4]':'County_Established', 
                   'Etymology[5]' : 'County_Info',
                   'Pop.(2021)[6]':'Population', 
                   'Area[4][7]' : 'Area_ToChange'
                 }
df.rename(columns = cols_to_rename, inplace=True)
df.head()
```

## 4. Split Column

```python
df.dtypes
```

```python
df.shape
```

```python
two_new_cols = ['County_Name', 'Delete']

df[two_new_cols] = df['County_Name_Full'].str.split(' ',1, expand=True)

print(df.shape)
df.head()
```

```python
df.drop("Delete", axis='columns', inplace=True)
df.head()
```

## 5. Changing data types

```python
df.dtypes
```

```python
two_new_cols = ['Area_SquareMiles', 'Delete']

df[two_new_cols] = df['Area_ToChange'].str.split('sq',1, expand=True)

print(df.shape)
df.head()
```

```python
df.drop(["Area_ToChange", "Delete"], axis='columns', inplace=True)
df.head()
```

```python
df['Area_SquareMiles'] = df['Area_SquareMiles'].str.replace(",", "")
```

```python
df['Area_SquareMiles'] = df['Area_SquareMiles'].astype(int)
```

```python
df.dtypes
```

```python
df.head()
```
