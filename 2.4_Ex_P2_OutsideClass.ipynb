# Q2.1:  Show the rows in North Carolina, Georgia and Virginia with Sales greater than $3,000 

- Sort by Segment (alphabetically),  Customer Name (alphabetically), and Sales (Greatest to Least)

```python
states = ['North Carolina', 'Georgia', 'Virginia']
df_Q1 = df.query("State in @states and Sales > 3000")
df_Q1.sort_values(by=['Segment', 'Customer Name', 'Sales'], ascending=[True, True, False], inplace=True)


print("Ex 2.4 Part 2: Show the rows in North Carolina, Georgia and Virginia with Sales greater than $3,000 ")
print(df_Q1.shape)
df_Q1
```
