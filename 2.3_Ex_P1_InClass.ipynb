# Q1.1:  How many Fords had an MPG Highway of at least 20 mpg? 

- Sort this dataframe by MPG Highway, Greatest to Least 


```python
df['Vehicle_Make'].unique()
```

```python
df_Q1 = df.query("Vehicle_Make == 'Ford' and MPG_Hwy >= 20")

print("Q1.1:  How many Fords had an MPG Highway of at least 20 mpg? ")
print(df_Q1.shape)
df_Q1.sort_values(by=['MPG_Hwy'], ascending=False)
```
