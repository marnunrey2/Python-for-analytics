# Q3: How Has the Pandemic Effected NYC Borough Rents? 

```python
df.drop("areaName", axis='columns', inplace=True)
df.drop("areaType", axis='columns', inplace=True)
df.head()
```

```python
df_Q3 = df.melt(id_vars=['Borough'], 
                  var_name="Date", 
                  value_name="Rent")

print(df_Q3.shape)
df_Q3.head()
```

```python
df_Q3.dtypes
```

```python
ser = df_Q3.groupby(["Borough","Date"])['Rent'].mean()
if type(ser) == pd.core.series.Series :
    print("It's a Series! Change it to a Dataframe!")
    df_Q3 = ser.to_frame()
    df_Q3.reset_index(inplace=True)
else:
    print("It's NOT a Series!")
    df_Q3 = ser
    
print(df_Q3.shape)
df_Q3.head()
```

```python
print("Part 3: How Has the Pandemic Effected NYC Borough Rents?")
print(df_Q3.shape)
df_Q3
```

```python
df_Q3.dtypes
```

### Plot

```python
fig = px.line(df_Q3, 
             x='Date', 
             y='Rent',
             color = 'Borough',
             template='ggplot2',
             title='Q3: How Has the Pandemic Effected NYC Borough Rents?'
             )

fig.show()
```
