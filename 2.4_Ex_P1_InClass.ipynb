# Q1.1:  What were the Top 5 Chevrolets for MPG Highway? 

- Sort this dataframe by MPG Highway (Greatest to Least) and Vehicle Model (Alphabetically) 

```python
df_Q1 = df.query("Vehicle_Make == 'Chevrolet'")
df_Q1.sort_values(by=['MPG_Hwy', 'Vehicle_Model'], ascending=[False, True], inplace=True)

df_TopN = df_Q1.iloc[ : 5]

print("Ex 2.4 Part 1")
print(df_TopN.shape)
df_TopN
```
