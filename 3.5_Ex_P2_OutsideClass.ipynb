# Ex 3.5: Part 2 (Outside Class)

```python
import pandas as pd
import plotly.express as px
```

### Read data  

```python
df = pd.read_excel('Data/MORTGAGE30US.xls', 'Slightly Cleaned', skiprows=0)

print(df.shape)
df.head()
```

```python
df.dtypes
```

# P2.1: Weekly 30 Year Fixed Mortgage Rate

```python
df_wkly = df.resample('W', on='Date').mean()

df_wkly.reset_index(inplace=True)

print(df_wkly.shape)
df_wkly.head()
```

```python
fig = px.line(df_wkly,              
             x='Date', 
             y='MortgageRate_30year_Fixed',
             template='plotly_dark',
             title='P2.1: Weekly 30 Year Fixed Mortgage Rate')

fig.update_xaxes(rangeslider_visible=True)

fig.show()
```
