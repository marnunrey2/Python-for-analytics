```python
import pandas as pd
import plotly.express as px
```

### Read data  

```python
df = pd.read_excel('Data/DEXCAUS.xls', 'Slightly Cleaned', skiprows=0)

print(df.shape)
df.head()
```

```python
df.dtypes
```

# P1.1: Daily Canada - US Exchange Rate

```python
fig = px.line(df,              
             x='Date', 
             y='US-Canada Exchange Rate',
             template='plotly_white',
             title='P1.1 Daily Canada - US Exchange Rate')

fig.update_xaxes(rangeslider_visible=True)

fig.show()
```

# P1.2: Monthly Average Canada - US Exchange Rate

```python
df_mtly = df.resample('M', on='Date').mean()

df_mtly.reset_index(inplace=True)

print(df_mtly.shape)
df_mtly.head()
```

```python
fig = px.line(df_mtly,              
             x='Date', 
             y='US-Canada Exchange Rate',
             template='gridon',
             title='P1.2 Monthly Average Canada - US Exchange Rate')

fig.update_xaxes(rangeslider_visible=True)

fig.show()
```
