# Part 2: Read xlsx  

```python
df = pd.read_excel("Data/World Bank 2020 Percent Internet Users.xlsx", skiprows=2)

print(df.shape)
df.head()
```

```python
df.dtypes
```

# Q2.1:  In 2020, what were the Top 20 Countries for % Internet Users? 

```python
df_Q1 = df.sort_values(by='Percent Internet Users', ascending=False)

print("Q2.1:  In 2020, what were the Top 20 Countries for % Internet Users?")

df_TopN = df_Q1.iloc[ : 20]
print(df_TopN.shape)
df_TopN
```

# Q2.2:  In 2020, what were the Bottom 20 Countries for % Internet Users? 

```python
df_Q2 = df.sort_values(by='Percent Internet Users', ascending=True)

print("Q2.2:  In 2020, what were the Bottom 20 Countries for % Internet Users? ")

df_TopN = df_Q2.iloc[ : 20]
print(df_TopN.shape)
df_TopN
```
