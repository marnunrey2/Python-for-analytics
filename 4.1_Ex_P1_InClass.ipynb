# 3. Create a pandas dataframe from the chosen list item of data

```python
df = pd.DataFrame(tables[0])  

print("Ex 4.1 Part 1")
print(df.shape)
df.head()
```

```python
df.dtypes
```

# 4. Cleaning dataframe

```python
df.drop("Traditionalabbreviation", axis='columns', inplace=True)

cols_to_rename = {
                   'State':'State_Name', 
                   'Otherabbreviations' : 'State_Abbrev'
                 }
df.rename(columns = cols_to_rename, inplace=True)
```

```python
df = df.query("State_Name != 'District of Columbia'")
```

```python
print("Ex 4.1 Part 1", df.shape)
df.head()
```

# 5. Save the Dataframe as a CSV file

```python
csv_file_to_create = "State_Abbreviations.csv"

filename_with_path = "Data/" + csv_file_to_create
df.to_csv(filename_with_path, index=False)
```
